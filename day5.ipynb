```python
def make_map(lines):
    out={}
    sorted_lines=sorted(lines,key=lambda x: int(x.split()[1]))
    for l in sorted_lines:
        d,s,l = l.split()
        out[range(int(s),int(s)+int(l))] = range(int(d),int(d)+int(l))
    last=0
    for k in out.copy():
        first=k[0]
        if first>last:
            out[range(last,first)]=range(last,first)
        last=k[-1]+1
    
    return out
```

```python
def maps_to(m,key):
    for k,v in m.items():
        if key in k:
            return v[k.index(key)]
    return key
```

```python
se2so=make_map(lines[3:51])
so2fe=make_map(lines[53:86])
fe2wa=make_map(lines[88:98])
wa2li=make_map(lines[100:112])
li2te=make_map(lines[114:137])
te2hu=make_map(lines[139:177])
hu2lo=make_map(lines[179:])
```

```python
ms=2**32
for s in seeds:
    ts=maps_to(hu2lo,maps_to(te2hu,maps_to(li2te,maps_to(wa2li,maps_to(fe2wa,maps_to(so2fe,maps_to(se2so,s)))))))
    if ts<ms: ms=ts
ms
```

```python
it = iter(seeds)
seeds_tuples=[*zip(it, it)]
```

```python
s=0
for a,b in seeds_tuples:
    s+=len(range(a,a+b))
s
```

```python
def map_to_range(m,in_ranges):
    maps=[]
    in_maps=[]
    for r in in_ranges:
        for k,v in m.items():
            tr=range(max(r[0], k[0]), min(r[-1], k[-1])+1)
            if tr:
                maps.append(v[slice(k.index(tr[0]),k.index(tr[-1])+1)])
                in_maps.append(range(r[r.index(tr[0])],r[r.index(tr[-1])]+1))
    return maps
```

```python
ms=2**32
for a,b in seeds_tuples:
    r=range(a,a+b)
    ranges=map_to_range(hu2lo,map_to_range(te2hu,map_to_range(li2te,map_to_range(wa2li,map_to_range(fe2wa,map_to_range(so2fe,map_to_range(se2so,[r])))))))
    v=min([i[0] for i in ranges])
    if v<ms: ms=v
v
```
